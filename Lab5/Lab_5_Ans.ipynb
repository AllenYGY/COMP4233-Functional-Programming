```ocaml
type unary = S;;
type peano = Nil | Cons of unary * peano;;

let zero = Nil;;
let one  = Cons(S, Nil);;
let two = Cons(S, one);;

(*  
Nil: peano
Cons: unary * peano: peano  one
*)

let successor x  = 
  Cons(S, x);;

let three = successor two;;

let predecessor = function 
  | Nil -> Nil
  | Cons (_, a) -> a;;

(* two = predecessor three;; *)

let rec add  x = function
  | Nil -> x
  | Cons(_, b') -> successor (add x b');;

(* zero = (add zero  zero);; *)
exception NotNat;;
let rec peano_of_nat (n:int)=
  if n < 0 then raise NotNat
  else if n = 0 then Nil
  else Cons(S, peano_of_nat (n-1));;

(* peano_of_nat -1;; *)
peano_of_nat 0;;
peano_of_nat 1;;
```

```ocaml
type int_or_float =
  | Int of int      
  | Float of float  

exception TypeError;;

let ( +.+ ) x y = 
  match x, y with
  | Int a, Int b -> Int (a + b)
  | Float a, Float b -> Float (a +. b)
  | _ -> raise TypeError;;
  (* | Int a, Float b -> raise TypeError
  | Float a, Int b -> raise TypeError;; *)

( +.+ ) (Int 1)  (Int 2);;
( +.+ ) (Float 1.0) (Float 2.0);;
(* ( +.+ ) (Int 1) (Float 2.);; *)
(Int 1) +.+ (Int 2);;
```
