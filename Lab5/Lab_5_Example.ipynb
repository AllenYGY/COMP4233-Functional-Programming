```ocaml
exception Not_Lower_Letter;;

let upper x = 
  if x >= 'a' && x <= 'z' then
    char_of_int (int_of_char x - int_of_char 'a' + int_of_char 'A')
  else
    raise Not_Lower_Letter;;

upper 'a';;
upper 'b';;
(* upper '1';; *)
```

```ocaml
let upper2 x = 
  if x >= 'a' && x <= 'z' then
    Some (char_of_int (int_of_char x - int_of_char 'a' + int_of_char 'A'))
  else
    None;;



upper2 'a';;
upper2 'b';;
upper2 '1';;

let lower2 x =
  if x >= 'A' && x <= 'Z' then
    Some (char_of_int (int_of_char x - int_of_char 'A' + int_of_char 'a'))
  else
    None;; 

exception Undefined;;

let surely x =
  match x with
  | Some x -> x
  | _ -> raise Undefined;;

let ch = 'c';;


ch= surely (lower2 (surely(upper2 ch)));;

'a' = ('a' |> upper2 |> surely  |> lower2 |> surely);;
(* 'A' = ('A' |> upper2 |> surely  |> lower2 |> surely);; *)
   
```

```ocaml
let compare a = 
  try a = (a |> upper2 |> surely  |> lower2 |> surely) with
  | Undefined -> false;;

compare 'a';;
compare 'A';;
```

```ocaml
type unary = S;;
type peano = Nil | Cons of unary * peano;;

let zero = Nil;;
let one  = Cons(S, Nil);;
let two = Cons(S, one);;

(*  
Nil: peano
Cons: unary * peano: peano  one
*)

let successor x  = 
  Cons(S, x);;

let three = successor two;;

let predecessor = function 
  | Nil -> Nil
  | Cons (_, a) -> a;;

two = predecessor three;;
```

```ocaml


let rec add  x = function
  | Nil -> x
  | Cons(_, b') -> successor (add x b');;

zero = (add zero  zero);;

```

```ocaml
let rec nat_of_peano = function
  | Nil -> 0
  | Cons(_, b') -> 1 + nat_of_peano b';;
```
