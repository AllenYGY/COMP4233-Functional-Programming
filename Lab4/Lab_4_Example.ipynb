```ocaml
type grade = A | B | C | D | F;;

let score_to_grade (s:int) : grade = 
  if s>=80 then A
    else if s>=60 then B
  else if s>=40 then C
    else if s>=20 then D
      else F;;
  
score_to_grade 85;;
```

```ocaml
let rec int_lst_min =function
    | [] -> 0
    | h::t -> 
      if t=[] then h
      else min h (int_lst_min t);;

let rec int_lst_min = function
    | [] -> 0
    | h::t [] -> h 
    | h :: t -> min h (int_lst_min t);;
    
int_lst_min scr_lst;;
```

```ocaml
let rec all_replace lst a b = 
  match lst with
  | [] -> []
  | h::t -> if h = a then b::(all_replace t a b) else h::(all_replace t a b);;


let lst =[2;3;2];;

let a =2;;
let b =4;;

all_replace lst a b;;
```

```ocaml
let rec last = function 
  | [] -> 0
  | [x] -> x
  | h::t -> last t;;

last lst;;

let rec last = function 
  | [] -> 0
  | h::[] ->h
  | h::t -> last t;;

last lst;;
```
