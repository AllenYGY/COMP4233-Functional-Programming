```ocaml
let rec map f =  function
  | [] -> [] 
  | h :: t -> f h :: map f t;;

let rec filter p = function
  | [] -> [] 
  | h :: t -> (if p h then [h] else [])@ filter p t;

let rec fold f ini = function
  | [] -> ini 
  | h :: t -> f h (fold f ini t);;
```

```ocaml
let scr_lst =92:: 81:: 75:: 64:: 55:: 49:: 37:: 23:: 11::[] ;;

type grade = A|B|C|D|F;;

let score_to_grade (s:int) : grade = 
  if s>=80 then A
    else if s>=60 then B
  else if s>=40 then C
    else if s>=20 then D
      else F;;

map score_to_grade scr_lst;;
(* val it : grade list = [A; A; B; B; C; C; D; F] *)
```

```ocaml
let fail score = 
  match score_to_grade score with
  | F -> true
  | _ -> false;;

filter fail scr_lst;;
```

```ocaml
let min a b = 
  if a < b then a else b;;
let max a b =
  if a > b then a else b;;

let min_scr = fold min 100 scr_lst;;
let max_scr = fold max 0 scr_lst;;
```

```ocaml
let currify f x y = f (x,y);;

let uncurrify f (x, y) = f x y;;
```

```ocaml
let id x = x;;

let apply f x = f x;;

let compose f g x= f (g x);;

let repose f g x = g (f x);;
```

```ocaml
type int_vector = int list;;

exception Different_dimension;;

let rec vector_operation f v1 v2 =
  match v1, v2 with
  | [], [] -> []
  | h1::t1, h2::t2 -> f h1 h2 :: vector_operation f t1 t2
  | _ -> raise Different_dimension;;

let vec_add a b = vector_operation ( + ) a b;; 
let convolution a b = vector_operation ( * ) a b;;

let rec bi_map f v1 v2 =
  match v1, v2 with
  | [], [] -> []
  | h1::t1, h2::t2 -> f h1 h2 :: bi_map f t1 t2
  | _ -> raise Different_dimension;;

let lst1= [1;2;3];; 
let lst2 = [4;5;6];;

vec_add lst1 lst2;;
```
